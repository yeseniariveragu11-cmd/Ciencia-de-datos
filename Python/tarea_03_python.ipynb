## 6. ¿Cuál es la diferencia entre una Tupla y una Lista?

La diferencia principal entre ambas es la mutabilidad:

Listas [ ]: Son mutables. Esto significa que podemos modificar, agregar o eliminar elementos después de que la lista ha sido creada. Se usan para colecciones de datos que cambian con el tiempo.

Tuplas ( ): Son inmutables. Una vez que se definen, su contenido no puede ser cambiado ni alterado. Funcionan como una "foto" fija de los datos para protegerlos de errores.

## 11. Función `domainGet`. Cree una función llamada domainGet que extraiga el dominio de un correo electrónico

Por ejemplo:
`domainGet("juan_perez@correo.com")` debería retornar 'correo.com'.

````python
def domainGet(email):
    # Partimos el texto en el símbolo '@'
    partes = email.split('@')
    
    # Retornamos el segundo elemento (posición 1) que es el dominio
    return partes[1]

correo_usuario = "yrivera085@soyudemedellin.edu.co"
dominio = domainGet(correo_usuario)

# 3. Imprimimos el resultado
print(f"El dominio del correo es: {dominio}")
````

## 12. Función `buscar_palabra_promedio`

Cree una función llamada `palabra_promedio` que retorne `True` si la palabra "promedio" (en cualquier forma de mayúsculas/minúsculas) está contenida en el string de entrada.

Ejemplo:

`palabra_promedio("El Promedio de la clase es de 4.3")` debería retornar `True`.

````python
def palabra_promedio(frase):
    # Convertimos toda la frase a minúsculas y preguntamos si 'promedio' está adentro
    return "promedio" in frase.lower()

resultado = palabra_promedio("El Promedio de la clase es de 4.3")
print(resultado)
````

## 13. Función contar_python

Cree una función llamada contar_python que cuente la cantidad de veces que aparece la palabra "python" en una cadena (ignorando mayúsculas y minúsculas).

Ejemplo:

`contar_python("Python es un lenguaje de programacion, Python es divertido")` debería retornar 2.

````python
def contar_python(frase):
    # 1. Convertimos toda la frase a minúsculas
    frase_limpia = frase.lower()
    
    # 2. Contamos cuántas veces aparece "python"
    cantidad = frase_limpia.count("python")
    
    return cantidad

resultado = contar_python("PYTHON es un lenguaje de programacion, Python es divertido")
print(f"La palabra 'python' aparece: {resultado} veces")
````

## 14. 💣 Detección de anomalías

Un científico de datos necesita desarrollar una función para detectar anomalías en un conjunto de datos de valores positivos. La función `deteccion_anomalia` debe seguir estas reglas:

1. Recibe los siguientes argumentos:

    * `valor (float):` El valor a evaluar.

    * `valor_media (float):` La media del conjunto de datos.

    * `desviacion_estandar (float):` La desviación estándar.

    * `distribucion_normal (bool):` Indica si la distribución es normal.

2. Si la distribución es **normal** y el valor es mayor a **valor_media + 3 * desviacion_estandar**, la función retorna **"anormal"**.

3. Si la distribución es **normal** y el valor es menor a **valor_media - 3 * desviacion_estandar**, la función retorna **"normal"**.

4. En cualquier otro caso, retorna **"indeterminado"**.

````python
# Copie su código acá

def deteccion_anomalia(valor, valor_media, desviacion_estandar, distribucion_normal):
    
    # Regla 2: Distribución normal y el valor es mayor a la media + 3 desviaciones
    if distribucion_normal == True and valor > (valor_media + 3 * desviacion_estandar):
        return "anormal"
    
    # Regla 3: Distribución normal y el valor es menor a la media - 3 desviaciones
    if distribucion_normal == True and valor < (valor_media - 3 * desviacion_estandar):
        return "normal"
    
    # Regla 4: Cualquier otro caso (si no se cumplen las anteriores)
    return "indeterminado"
````

````python
# Ejemplos de prueba

print(deteccion_anomalia(
    valor=15.6, valor_media=9.5, desviacion_estandar=8.7, distribucion_normal=True
))  # Se espera: 'normal'

print(deteccion_anomalia(
    valor=45.4, valor_media=9.5, desviacion_estandar=8.7, distribucion_normal=True
))  # Se espera: 'anormal'

print(deteccion_anomalia(
    valor=15.6, valor_media=9.5, desviacion_estandar=8.7, distribucion_normal=False
))  # Se espera: 'indeterminado'
````
