**¿Cuántas personas tienen la palabra Chief en el título de su trabajo?** Apoyese en los métodos para manipular strings y en el método `.apply()`

```python
# Definimos la función de búsqueda
def tiene_chief(titulo):
    # Convertimos a minúsculas para que no importe cómo esté escrito
    if 'chief' in titulo.lower():
        return True
    else:
        return False

# Aplicamos la función a la columna y sumamos los True
cantidad_chiefs = salarios['JobTitle'].apply(tiene_chief).sum()

# Imprimimos el resultado
print(f"Hay {cantidad_chiefs} personas que tienen la palabra 'Chief' en su título de trabajo.")
```

Este código es valido, se recomienda hacer el código mas simple y agregar typing y documentación.

```python
def buscar_palabra_en_puestos(df: pd.DataFrame, palabra: str) -> int:
    """
    Cuenta cuántos registros en la columna 'JobTitle' contienen una palabra específica.
    
    Args:
        df (pd.DataFrame): El DataFrame que contiene los datos de salarios.
        palabra (str): La palabra clave a buscar (ej. 'chief').
        
    Returns:
        int: El número total de personas que tienen esa palabra en su cargo.
    """
    # Usamos str.contains con case=False para ignorar mayúsculas/minúsculas
    # y na=False para evitar errores si hay celdas vacías.
    conteo = df['JobTitle'].str.contains(palabra, case=False, na=False).sum()
    return int(conteo)

# Ejecutamos la función y guardamos el resultado
cantidad_chiefs = buscar_palabra_en_puestos(salarios, 'chief')

# Mostramos el resultado final
print(f"Resultado: {cantidad_chiefs} personas tienen la palabra 'Chief' en su título de trabajo.")
```
