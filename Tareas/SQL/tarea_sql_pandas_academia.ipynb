## 2. Convertir a DataFrames

```python
estudiantes = pd.DataFrame(estudiantes_data, columns=["estudiante_id", "nombre", "ciudad", "fecha_registro"])
cursos = pd.DataFrame(cursos_data, columns=["curso_id", "nombre_curso", "categoria", "precio", "nivel"])
inscripciones = pd.DataFrame(inscripciones_data, columns=["inscripcion_id", "estudiante_id", "curso_id", "fecha_inscripcion", "estado"])
pagos = pd.DataFrame(pagos_data, columns=["pago_id", "inscripcion_id", "monto", "metodo_pago", "fecha_pago"])

estudiantes["fecha_registro"] = pd.to_datetime(estudiantes["fecha_registro"])
inscripciones["fecha_inscripcion"] = pd.to_datetime(inscripciones["fecha_inscripcion"])
pagos["fecha_pago"] = pd.to_datetime(pagos["fecha_pago"])

display(estudiantes.head())
display(cursos.head())
display(inscripciones.head())
display(pagos.head())
```

## 3. Cargar tablas a SQLite

```python
conn = sqlite3.connect(":memory:")

estudiantes.to_sql("estudiantes", conn, index=False, if_exists="replace")
cursos.to_sql("cursos", conn, index=False, if_exists="replace")
inscripciones.to_sql("inscripciones", conn, index=False, if_exists="replace")
pagos.to_sql("pagos", conn, index=False, if_exists="replace")

def run_sql(query):
    return pd.read_sql_query(query, conn)
```

### Punto 1. Selección y filtro básico

Trae todos los estudiantes que viven en **Bogotá**.

Trae los cursos ordenados de **mayor a menor precio** y muestra solo:

### Punto 3. Filtro con múltiples condiciones

Trae las inscripciones cuyo estado sea **activo** y cuya fecha de inscripción sea posterior al **2024-04-15**.

### Punto 4. Agregación simple

Calcula:
- número total de cursos
- precio promedio
- precio mínimo
- precio máximo

Usa la tabla `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 4 - SQL
query_4 = """
SELECT 
    COUNT(curso_id) AS total_cursos,
    AVG(precio) AS precio_promedio,
    MIN(precio) AS precio_minimo,
    MAX(precio) AS precio_maximo
FROM cursos;
"""

resultado_sql_4 = run_sql(query_4)

display(resultado_sql_4)
```

```python
# Punto 4 - pandas
resultado_pandas_4 = pd.DataFrame({
    'total_cursos': [cursos['curso_id'].count()],
    'precio_promedio': [cursos['precio'].mean()],
    'precio_minimo': [cursos['precio'].min()],
    'precio_maximo': [cursos['precio'].max()]
})

display(resultado_pandas_4)
```

### Punto 5. GROUP BY

Cuenta cuántos cursos hay por `categoria`.

Ordena de mayor a menor cantidad.

**Entregable:**
- SQL
- pandas

```python
# Punto 5 - SQL
query_5 = """
SELECT 
    categoria, 
    COUNT(curso_id) AS cantidad_cursos
FROM cursos
GROUP BY categoria
ORDER BY cantidad_cursos DESC;
"""

resultado_sql_5 = run_sql(query_5)

display(resultado_sql_5)
```

```python
# Punto 5 - pandas
resultado_pandas_5 = (
    cursos.groupby('categoria')['curso_id']
    .count()
    .reset_index(name='cantidad_cursos')
    .sort_values(by='cantidad_cursos', ascending=False)
)

display(resultado_pandas_5)
```

### Punto 6. JOIN entre tablas

Muestra:
- nombre del estudiante
- nombre del curso
- estado de la inscripción

Usa `estudiantes`, `inscripciones` y `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 6 - SQL
query_6 = """
SELECT 
    estudiantes.nombre, 
    cursos.nombre_curso, 
    inscripciones.estado
FROM inscripciones
JOIN estudiantes ON inscripciones.estudiante_id = estudiantes.estudiante_id
JOIN cursos ON inscripciones.curso_id = cursos.curso_id;
"""

resultado_sql_6 = run_sql(query_6)

display(resultado_sql_6)
```

```python
# Punto 6 - pandas
union_1 = pd.merge(inscripciones, estudiantes, on='estudiante_id')

union_2 = pd.merge(union_1, cursos, on='curso_id')

resultado_pandas_6 = union_2[['nombre', 'nombre_curso', 'estado']]

display(resultado_pandas_6)
```

### Punto 7. LEFT JOIN

Trae todos los estudiantes, incluso si no tuvieran inscripciones, mostrando:
- `estudiante_id`
- `nombre`
- `inscripcion_id`

**Entregable:**
- SQL
- pandas

> Pista: en este dataset puede que todos tengan al menos una inscripción, pero igual debes practicar el `LEFT JOIN`.

```python
# Punto 7 - SQL
query_7 = """
SELECT 
    estudiantes.estudiante_id, 
    estudiantes.nombre, 
    inscripciones.inscripcion_id
FROM estudiantes
LEFT JOIN inscripciones ON estudiantes.estudiante_id = inscripciones.estudiante_id;
"""

resultado_sql_7 = run_sql(query_7)

display(resultado_sql_7)
```

```python
# Punto 7 - pandas
union_left = pd.merge(estudiantes, inscripciones, on='estudiante_id', how='left')

resultado_pandas_7 = union_left[['estudiante_id', 'nombre', 'inscripcion_id']]

display(resultado_pandas_7)
```

### Punto 8. HAVING

Calcula cuántas inscripciones tiene cada estudiante y muestra solo aquellos con **2 o más**.

Debes mostrar:
- `estudiante_id`
- `nombre`
- `n_inscripciones`

**Entregable:**
- SQL
- pandas

```python
# Punto 8 - SQL
query_8 = """
SELECT 
    e.estudiante_id, 
    e.nombre, 
    COUNT(i.inscripcion_id) AS n_inscripciones
FROM estudiantes e
JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
GROUP BY e.estudiante_id, e.nombre
HAVING COUNT(i.inscripcion_id) >= 2;
"""

resultado_sql_8 = run_sql(query_8)

display(resultado_sql_8)
```

```python
# Punto 8 - pandas
union_8 = pd.merge(estudiantes, inscripciones, on='estudiante_id')

agrupado_8 = (
    union_8.groupby(['estudiante_id', 'nombre'])['inscripcion_id']
    .count()
    .reset_index(name='n_inscripciones')
)
resultado_pandas_8 = agrupado_8[agrupado_8['n_inscripciones'] >= 2]

display(resultado_pandas_8)
```

### Punto 9. CASE WHEN / segmentación

Clasifica los cursos según su precio:
- `Bajo` si el precio es menor a 120
- `Medio` si el precio está entre 120 y 180 inclusive
- `Alto` si el precio es mayor a 180

Debes mostrar:
- `nombre_curso`
- `precio`
- `segmento_precio`

**Entregable:**
- SQL
- pandas

```python
# Punto 9 - SQL
query_9 = """
SELECT 
    nombre_curso,
    precio,
    CASE 
        WHEN precio < 120 THEN 'Bajo'
        WHEN precio >= 120 AND precio <= 180 THEN 'Medio'
        WHEN precio > 180 THEN 'Alto'
    END AS segmento_precio
FROM cursos;
"""

resultado_sql_9 = run_sql(query_9)

display(resultado_sql_9)
```

```python
# Punto 9 - pandas
resultado_pandas_9 = cursos[['nombre_curso', 'precio']].copy()

resultado_pandas_9.loc[resultado_pandas_9['precio'] < 120, 'segmento_precio'] = 'Bajo'
resultado_pandas_9.loc[(resultado_pandas_9['precio'] >= 120) & (resultado_pandas_9['precio'] <= 180), 'segmento_precio'] = 'Medio'
resultado_pandas_9.loc[resultado_pandas_9['precio'] > 180, 'segmento_precio'] = 'Alto'

display(resultado_pandas_9)
```

### Punto 10. Ejercicio final integrador

Calcula el total pagado por categoría de curso.

Debes:
- unir `cursos`, `inscripciones` y `pagos`
- agrupar por `categoria`
- sumar `monto`
- ordenar de mayor a menor

**Entregable:**
- SQL
- pandas

```python
# Punto 10 - SQL
query_10 = """
SELECT 
    c.categoria, 
    SUM(p.monto) AS total_pagado
FROM cursos c
JOIN inscripciones i ON c.curso_id = i.curso_id
JOIN pagos p ON i.inscripcion_id = p.inscripcion_id
GROUP BY c.categoria
ORDER BY total_pagado DESC;
"""

resultado_sql_10 = run_sql(query_10)

display(resultado_sql_10)
```

```python
# Punto 10 - pandass
union_1 = pd.merge(cursos, inscripciones, on='curso_id')
union_total = pd.merge(union_1, pagos, on='inscripcion_id')

resultado_pandas_10 = (
    union_total.groupby('categoria')['monto']
    .sum()
    .reset_index(name='total_pagado')
    .sort_values(by='total_pagado', ascending=False)
)
display(resultado_pandas_10)
```
