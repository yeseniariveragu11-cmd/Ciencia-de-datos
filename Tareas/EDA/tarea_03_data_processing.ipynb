Trabajarás con el dataset del Titanic para practicar transformaciones reales.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

---
## Parte 1 — Exploración y selección

**P1.** Imprime los nombres de todas las columnas y el número total de filas y columnas del DataFrame.

```python
print(df.columns.tolist())
print(df.shape)
```

**P2.** Usa `.loc` para seleccionar las filas 10 a 15 (inclusive) y solo las columnas `Name`, `Age` y `Survived`.

```python
seleccion = df.loc[10:15, ["Name", "Age", "Survived"]]

print(seleccion)
```

---
## Parte 2 — Valores nulos y limpieza

**P3.** Cuenta cuántos valores nulos hay en cada columna. ¿Cuáles columnas tienen nulos?

```python
nulos_por_columna = df.isnull().sum()

print("Cantidad total de nulos por columna:")
print(nulos_por_columna)

print("\n---")

columnas_con_nulos = nulos_por_columna[nulos_por_columna > 0]

print("Las columnas que tienen valores nulos son:")
print(columnas_con_nulos)
```

**Respuesta:** Las columnas con valores nulos son: Age, Cabin y Embarked

**P4.** Rellena los valores nulos de la columna `Age` con la **mediana** de esa columna.  
Verifica que ya no haya nulos en `Age`.

```python
mediana_edad = df['Age'].median()
print("La mediana de edad es:", mediana_edad)

df['Age'] = df['Age'].fillna(mediana_edad)

nulos_edad_ahora = df['Age'].isnull().sum()

print("Cantidad de nulos en Age después de rellenar:", nulos_edad_ahora)
```

---
## Parte 3 — Nuevas columnas y agrupaciones

**P5.** Crea una nueva columna llamada `FamilySize` que sea la suma de `SibSp` + `Parch` + 1 (el pasajero mismo).

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

print(df[['SibSp', 'Parch', 'FamilySize']].head())
```

**P6.** Usando `groupby`, calcula la **tasa de supervivencia promedio** (columna `Survived`) por clase (`Pclass`).  
¿Qué clase tuvo mayor probabilidad de sobrevivir?

```python
tasa_supervivencia = df.groupby('Pclass')['Survived'].mean()

print(tasa_supervivencia)
```

**Respuesta:** La clase con mayor supervivencia fue:  La primera clase (1)

**P7.** Calcula la edad promedio y la tarifa (`Fare`) promedio, agrupando por `Sex` y `Pclass`.  
Usa `.groupby()` con `.agg()`.

```python
agrupacion = df.groupby(['Sex', 'Pclass'])

promedios = agrupacion.agg({
    'Age': 'mean', 
    'Fare': 'mean'
})

print(promedios)
```

## Parte 4 — Ordenamiento y duplicados

**P8.** Ordena el DataFrame por `Fare` de mayor a menor y muestra las primeras 5 filas.  
¿Cuánto pagó el pasajero más caro?

```python
df_ordenado = df.sort_values(by='Fare', ascending=False)

print(df_ordenado[['Name', 'Pclass', 'Fare']].head(5))
```

**Respuesta:** Hubo 3 pasajeros que pagaron el pasaje más caro que fue por un valor de 512.3292

**P9.** Verifica si hay filas duplicadas en el DataFrame. ¿Cuántas hay?

```python
cantidad_duplicados = df.duplicated().sum()

print(cantidad_duplicados)
```

**Respuesta:** No hay filas duplicadas
